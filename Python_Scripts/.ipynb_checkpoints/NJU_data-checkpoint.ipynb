```python
from math import sqrt
j=2
while j<=100:
    i=2
    k=sqrt(j)
    while (i<=k):
        if j%i==0: break
        i +=1
    if(i>k):
        print(j)
    j +=1    
```

```python
from math import sqrt
j=2
def is_prime(x):
    if x==1: return False
    k=int(sqrt(x))
    for j in range(2,k+1):
        if x%j==0: return False
    return True

for i in range(2, 101):
    if is_prime(i):
        print(i)    
```

```python
# recursion for fibonacci
def fib(n):
    if n==0 or n==1:
        return n
    else:
        return(fib(n-1)+fib(n-2))
fib(5)
```

```python
# recursion for fibonacci
def hanoi(a,b,c,n): # 目标是把盘子从A柱子， 借助B柱子，挪动到C柱子
    if n==1:
       print(a, "=>", c)
    else: 
        hanoi(a,c,b,n-1)
        print(a, "=>", c)
        hanoi(b,a,c,n-1)
       
hanoi('A','B','C', 2)
```
